# Performing Continuous Wavelet transform using scipy.signal package

### Here using Ricker wavelet as it was built-in Scipy signal package

```python
from scipy import signal
```

# Monsoon

```python
widths = np.arange(0.5,80,0.01) # Widths of wavelets to be used while performing the cwt 
# np.arange (start , stop , step)  higher the step, higher the resolution, slower the computation (for cwt)

cwtmonsoonJune = signal.cwt(monsoon[:,0], signal.ricker, widths)
cwtmonsoonJuly = signal.cwt(monsoon[:,1], signal.ricker, widths)
cwtmonsoonAug = signal.cwt(monsoon[:,2], signal.ricker, widths)
cwtmonsoonSep = signal.cwt(monsoon[:,3], signal.ricker, widths)
```

```python
plt.imshow(np.flipud(cwtmonsoonJune),interpolation ='bicubic', extent=[0, 70, 1, 70])
plt.title("June")
plt.show()
plt.imshow(np.flipud(cwtmonsoonJuly),interpolation ='bicubic', extent=[0, 70, 1, 70])
plt.title("July")
plt.show()
plt.imshow(np.flipud(cwtmonsoonAug),interpolation ='bicubic', extent=[0, 70, 1, 70])
plt.title("August")
plt.show()
plt.imshow(np.flipud(cwtmonsoonSep),interpolation ='bicubic', extent=[0, 70, 1, 70])
plt.title("September")
plt.show()
```

# Plots Shown are in Form = Y - Timeline, X - Widths as per Scipy.signal.cwt package

# In all above plots we can see that there exist a wavelet of width between 30 & 40 of Ricker in timeline starting from around 35 to 70 years. 

### Continuous Wavelet Transform with scipy package in python doesn't help with visualization as much as MATLAB does. Therefore, it is best to perform cwt in matlab for visualization

# Excersice 2 

```python
temp_data = pd.read_excel("Annual average temp.xlsx")
temp_data
```

```python
widths = np.arange(0.5,150,0.01)
avg_temp = signal.cwt(temp_data[temp_data.columns[1]], signal.ricker, widths)

plt.imshow(np.flipud(avg_temp),interpolation ='bicubic', extent=[0, 147, 1, 147])
plt.title("Average Temperature")
plt.show()
```

# Plots Shown are in Form = Y - Timeline, X - Widths as per Scipy.signal.cwt package

# From above plot we can say that a width of 60 to 80 of Ricker wavelet exists for time starting from around 60 to 140 years
