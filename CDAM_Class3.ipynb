```python
import numpy as np
from scipy.fft import rfft
import matplotlib.pyplot as plt
```

```python
x = np.arange(100)
y = np.sin(x) + np.sin(2*x) + np.sin(3*x)
plt.plot(y)
plt.show()
y_fft = rfft(y)
plt.plot(np.abs(y_fft))
plt.show()
```

```python
y_0 = np.array(())
y_1 = np.array(())
y_2 = np.array(())

x_0 = np.arange(100)
x_1 = np.arange(100, 200)
x_2 = np.arange(200, 300)

y_0 = np.sin(x_0)
y_1 = np.sin(2*x_1)
y_2 = np.sin(3*x_2)

plt.plot(x_0, y_0, x_1, y_1, x_2, y_2)
plt.show()
y = np.concatenate((y_0, y_1, y_2))

y_fft = rfft(y)

plt.plot(np.abs(y_fft))

plt.show()
```
