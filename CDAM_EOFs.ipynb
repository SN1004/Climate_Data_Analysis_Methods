```python
var = ds['Long']

# Perform EOF analysis
solver = Eof(var)

pcs = solver.pcs(npcs=3, pcscaling=1)
eofs = solver.eofs(neofs=3)

# Plot the first EOF
fig, ax = plt.subplots()
eofs.sel(mode=0).plot(ax=ax)
ax.set(title='First EOF')

# Plot the time series of the first PC
fig, ax = plt.subplots()
pcs.sel(mode=0).plot(ax=ax)
ax.set(title='First PC')
var
```
